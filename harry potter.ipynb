```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("books.csv")
ds
```

```python
ds.columns
```

```python
null_values = ds.isnull().sum()

# Display the number of null values
null_values
```

```python
cleaned_ds = ds.dropna()
```

```python
cleaned_ds
```

```python
null_values = cleaned_ds.isnull().sum()
null_values
```

```python
cleaned_ds.drop(columns=['small_image_url', 'image_url', 'original_title', 'book_id','goodreads_book_id','best_book_id','work_id','isbn','isbn13'], inplace=True)
```

```python
cleaned_ds
```

```python
# Basic statistics
cleaned_ds.describe()
```

```python
#Calculate total ratings count for each book
cleaned_ds['total_ratings'] = cleaned_ds['ratings_1'] + cleaned_ds['ratings_2'] + cleaned_ds['ratings_3'] + cleaned_ds['ratings_4'] + cleaned_ds['ratings_5']

# Sort the dataset by total ratings in descending order
most_selling_books = cleaned_ds[['title', 'total_ratings']].sort_values(by='total_ratings', ascending=False)

print("Top 5 most selling books within the Harry Potter series:")
print(most_selling_books.head(20))
```

```python
# Calculate the average rating of the Harry Potter books
average_rating = cleaned_ds['average_rating'].mean()
print("\nAverage rating of the Harry Potter books:", average_rating)
```
